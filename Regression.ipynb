```python
X_data = np.load('x_test.npy')
Y_data = np.load('y_test.npy')
XT = np.load('x_train.npy')
YT = np.load('y_train.npy')
print(XT, YT)
```

```python
xt = XT[:, 0]
features = np.array([xt**3, xt**2, xt, np.ones_like(xt)]).T
print(features)
```

```python

def Loss_function(W, features, YT):
    loss = 0
    grad = np.zeros(4)              #change input for zeros base on number of features for the linear regression
    for x, y in zip(features, YT):
        y_model = np.sum(W * x)
        loss += (y - y_model)**2
        grad += -2 * x * (y - y_model)   
    return loss, grad 

W = np.array([0, 1, 1, 0])          #Change input base on number of features for the linear regression
Loss_function(W, features, YT)
```

```python
loss = 1000
W = np.array([0, 1, 1, 0])
alpha = 0.00000001
while loss > 0.0001:
    loss, grad = Loss_function(W, features, YT)
    print(loss, grad, W)
    W = W - alpha * grad
```
